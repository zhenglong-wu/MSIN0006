```python
import pandas as pd

data = pd.read_csv('raw_data.csv')
data.set_index('date', inplace=True)
data.head()
```

```python
new = pd.DataFrame()
for column in data:
    if column == 'harrypotterwiki':
        new = pd.DataFrame(data['harrypotterwiki'], index=data.index)
        new['source'] = column
        new.rename(columns={'harrypotterwiki': 'searches'}, inplace=True)
    else:
        temp = pd.DataFrame(data[column], index=data.index)
        temp['source'] = column
        temp.rename(columns={f'{column}': 'searches'}, inplace=True)
        pd.concat([new, temp], axis=1)

print(len(new))
new.tail()
```
